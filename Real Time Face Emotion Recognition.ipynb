## Real time video demo for Face Emotion Recognition

```python
# import dependecies
import cv2
from deepface import DeepFace
import matplotlib.pyplot as plt
```

```python
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
def list_to_dict(predictions):
    # Extracting column names
    column_names = list(predictions[0].keys())

    # Creating a dictionary with column names as keys
    data_dict = {column: [entry[column] for entry in predictions] for column in column_names}

    # return the resulting dictionary
    return data_dict
```

```python
import cv2
from deepface import DeepFace

faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

cap = cv2.VideoCapture(1)

#check the webcam is opened correctly
if not cap.isOpened():
    cap = cv2.VideoCapture(0)

if not cap.isOpened():
    raise IOError('Cannot open webcam')

while True:
    ret, frame = cap.read()
    result = DeepFace.analyze(frame, actions=['emotion'], enforce_detection=False)
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    faces = faceCascade.detectMultiScale(gray, 1.1, 4)

    # Draw Rectangle around the face
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x,y), (x+w, y+h), (0, 255, 0), 2)
    
    font = cv2.FONT_HERSHEY_SIMPLEX
    
    # list_to_dict
    data_dict = list_to_dict(result)
    
    # Stressed or not Stressed
    if data_dict['dominant_emotion'][0] in ['scared','sad', 'surprise', 'fear']:
        label = 'stressed'
    else:
        label = 'not stressed'
    
    # USE putText method for inserting text on video
    cv2.putText(
        frame,
            label,
            (50, 50),
            font, 3,
            (0, 0, 255),
            2,
            cv2.LINE_4);
    cv2.imshow("Original video", frame)
    
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

# Real Time Emotion Recognition using Tensorflow

## read all images and convert them to an array

```python
training_data = []

def create_training_data():
    for category in Classes:
        path = os.path.join(DataDirectory, category)
        class_name = category
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path, img))
                new_array = cv2.resize(img_array, (img_size,img_size))
                training_data.append([new_array, class_name])
            except Exception as e:
                pass
```

```python
create_training_data()
```

```python
print(len(training_data))
```

```python
import random

random.shuffle(training_data)
```

```python
#train_data = training_data[:10001]
```

```python
#img_size = 224
```

```python
X = [] # data/feature
y = [] # label

for features, label in training_data:
    X.append(features)
    y.append(label)

X= np.array(X).reshape(-1, img_size,img_size,3) # convet it to 4 dimensions
```

```python
X.shape
```

```python
Y = np.array(y)
Y.shape
```

```python
X=X[:900]
Y=Y[:900]
```

```python
Y[1]
```

```python
X=X.astype('float32')
Y=Y.astype('int32')
```

```python
# normalize the data
X = X/255.0
```

```python
print(X.shape)
print(Y.shape)
```

# Deep learning model for training - Transfer learning

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = tf.keras.applications.MobileNetV2()
```

```python
model.summary()
```

## Transfer Learning - Tuning, weights will start from last check point

```python
base_input = model.layers[0].input
```

```python
base_output = model.layers[-2].output
```

```python
final_output = layers.Dense(128)(base_output)
final_output = layers.Activation('relu')(final_output)
final_output = layers.Dense(68)(base_output)
final_output = layers.Activation('relu')(final_output)
final_output = layers.Dense(7, activation='softmax')(final_output) # we have 7 classes
```

```python
final_output
```

```python
new_model = keras.Model(inputs = base_input, outputs=final_output)
```

```python
new_model.summary()
```

```python
new_model.compile(loss='sparse_categorical_crossentropy' ,optimizer='adam', metrics=['accuracy'])
```

```python
new_model.fit(X,Y,epochs=25)
```

```python
new_model.save('my_model_001.h5')
```

```python
new_model = tf.keras.models.load_model('my_model_001.h5')
```

```python
img = cv2.imread('disgust_women.jpg')
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
faces = faceCascade.detectMultiScale(gray,1.1,4)

for x,y,w,h in faces:
    roi_gray = gray[y:y+h,x:x+w]
    roi_color = img[y:y+h,x:x+w]
    cv2.rectangle(img, (x,y), (x+w, y+h), (255, 0, 0), 2)
    facess = faceCascade.detectMultiScale(roi_gray)
    if len(facess) == 0:
        print('Face not detected')
    else:
        for (ex,ey,ew,eh) in facess:
            face_roi = roi_color[ey:ey+eh, ex:ex+ew]
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(face_roi, cv2.COLOR_BGR2RGB))
```

```python
img_size=224
```

```python
final_image = cv2.resize(face_roi, (224,224))
final_image = np.expand_dims(final_image, axis=0)
#final_image= np.array(final_image).reshape(-1, img_size,img_size,3) # convet it to 4 dimensions
final_image = final_image/255.0
```

```python
prediction = new_model.predict(final_image)
```
